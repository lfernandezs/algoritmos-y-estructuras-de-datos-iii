# Demostración empírica del problema de selección de actividades

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos los datos
df = pd.read_csv('output_unsorted.csv')

# Visualizamos los datos
df.head()

# Graficamos los datos
sns.scatterplot(x='time', y='n', data=df)
plt.xlabel('Tiempo (t)')
plt.ylabel('Tamaño (n)')
```

```python
# Cargamos los datos
df = pd.read_csv('output_sorted.csv')

# Visualizamos los datos
df.head()

# Graficamos los datos
sns.scatterplot(x='time', y='n', data=df)
plt.xlabel('Tiempo (t)')
plt.ylabel('Tamaño (n)')
```

```python
# Cargamos los datos
df_unsorted = pd.read_csv('output_unsorted.csv')
df_sorted = pd.read_csv('output_sorted.csv')

df_unsorted['sorted'] = False
df_sorted['sorted'] = True

df = pd.concat([df_unsorted, df_sorted])
sns.lmplot(x='time', y='n', data=df, hue='sorted', fit_reg=False, legend=False)

```
